```python
import pandas as pd
```

```python
df=pd.read_csv("50_Startups.csv")
```

```python
df
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df
```

```python
df.columns
```

```python
x_independent = df[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
y_dependent = df[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_independent,y_dependent,test_size = 0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
reg.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(reg)
plt.show()
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```

```python
# loop with all hyper parameter
# hyper parameter criterion
score_df=pd.DataFrame(columns=['criterion','splitter','r_score'])
criterion = ['squared_error', 'friedman_mse', 'absolute_error', 'poisson']
splitter =['best', 'random']
for i in criterion:
    for j in splitter:
        reg=DecisionTreeRegressor(criterion=i,splitter=j)
        reg.fit(x_train,y_train)
        y_pred = reg.predict(x_test)
        r_score = r2_score(y_test,y_pred)
        score_df.loc[len(score_df.index)] = [ i , j ,r_score]
score_df
```
